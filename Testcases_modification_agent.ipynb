```python
tools_list = [
    ReadFileTool(),
    WriteFileTool(),
    ListDirectoryTool(),
    CopyFileTool(),
    MoveFileTool(),
    DeleteFileTool(),
    repl_tool
]
```

```python
prog_name_list=os.listdir("python_programs")
new=[]
new1=[]
for i in prog_name_list:
    new.append(i[ :-3])
for x in new:

    if x[-4: ]=="test":
        continue
    else:
        new1.append(x)
prog_name_list=new1
prog_name_list=prog_name_list[14: ]
prog_name_list=prog_name_list[14: ]
prog_name_list=prog_name_list[9: ]
prog_name_list=prog_name_list[1: ]
prog_name_list
```

```python
agent_correction = initialize_agent(
    tools=tools_list,
    llm=llm,
    agent=AgentType.STRUCTURED_CHAT_ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
agent_correction_executor = AgentExecutor.from_agent_and_tools(
    agent=agent_correction.agent,
    tools=tools_list,
    verbose=True,
    max_iterations=1000,           # Increase as needed
    max_execution_time=3000      # 5 minutes
)
```
